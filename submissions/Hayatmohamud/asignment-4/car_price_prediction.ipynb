1. Import Libraries

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
```

2.Load Dataset

```python
import pandas as pd

df = pd.read_csv("car_l3_clean_ready.csv")

print("Dataset Shape:", df.shape)
print("\nFirst Five Rows")
print(df.head())
```

3.Prepare Features & Target

```python
# Target = Price
y = df["Price"]


X = df.drop(columns=["Price", "LogPrice"])

print("\nFeatures Shape:", X.shape)
```

4.Split Dataset

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("\nTraining Samples:", len(X_train))
print("Testing Samples :", len(X_test))
```

5.Train Linear Regression

```python
lr = LinearRegression()

lr.fit(X_train, y_train)

lr_pred = lr.predict(X_test)
```

6. Train Random Forest

```python
rf = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf.fit(X_train, y_train)

rf_pred = rf.predict(X_test)
```

7. Helper Function

```python
def print_metrics(model_name, y_true, y_pred):

    r2 = r2_score(y_true, y_pred)

    mae = mean_absolute_error(y_true, y_pred)

    mse = mean_squared_error(y_true, y_pred)

    rmse = np.sqrt(mse)

    print(f"\n{model_name} Performance:")
    print(f"  R²   : {r2:.3f}")
    print(f"  MAE  : {mae:,.2f}")
    print(f"  MSE  : {mse:,.2f}")
    print(f"  RMSE : {rmse:,.2f}")
```

8. Evaluate Both Models

```python
print_metrics(
    "Linear Regression",
    y_test,
    lr_pred
)

print_metrics(
    "Random Forest",
    y_test,
    rf_pred
)
```

9. Compare Predictions

```python
results = pd.DataFrame({

    "Actual Price": y_test.values,

    "Linear Prediction": lr_pred,

    "Random Forest Prediction": rf_pred

})

print("\nPrediction Results")
print(results.head(10))
```

10. Sanity Check

```python
i = 0

sample = X_test.iloc[[i]]

actual_price = y_test.iloc[i]

lr_price = lr.predict(sample)[0]

rf_price = rf.predict(sample)[0]

print("\nSanity Check")
print("----------------------------")
print(f"Actual Price            : {actual_price:,.2f}")
print(f"Linear Regression Price : {lr_price:,.2f}")
print(f"Random Forest Price     : {rf_price:,.2f}")
```
